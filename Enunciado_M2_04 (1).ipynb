* Realiza una búsqueda de un elemento dentro de la lista y de la tupla. Mostrar si devuelve True o False

* Borra o elimina (si se puede) el contenido de la lista y de la tupla. Mosstrar de nuevo la lista y la tupla para verificar si se ha realizado correctamente la acción



```python
lista_generos = ["Rock", "Jazz", "Reggae"]
tupla_generos = ("Pop", "Clásica", "Hip-Hop")

print("Lista de géneros:", lista_generos)
print("Tupla de géneros:", tupla_generos)

print(f"\n2º elemento de la lista: {lista_generos[1]}")
print(f"Penúltimo elemento de la tupla: {tupla_generos[-2]}")

print("\n--- Intentando modificar elementos ---")
lista_generos[0] = "Metal"
print(f"Lista después de modificar el 1er elemento: {lista_generos}")

print("Tuplas no se pueden modificar (son inmutables)")

print(f"\nTamaño de la lista: {len(lista_generos)}")
print(f"Tamaño de la tupla: {len(tupla_generos)}")

elemento_busqueda = "Jazz"
print(f"\n¿'{elemento_busqueda}' está en la lista? {elemento_busqueda in lista_generos}")
print(f"¿'{elemento_busqueda}' está en la tupla? {elemento_busqueda in tupla_generos}")

print("\n--- Añadiendo elementos ---")
lista_generos.append("Blues")
print(f"Lista después de añadir 'Blues': {lista_generos}")

tupla_generos = tupla_generos + ("Electrónica",)
print(f"Tupla después de añadir 'Electrónica': {tupla_generos}")

print("\n--- Eliminando contenido ---")
lista_generos.clear()
print(f"Lista después de limpiarla: {lista_generos}")

tupla_generos = ()
print(f"Tupla después de vaciarla: {tupla_generos}")
```
